```python
def get_ageb(long,lat):
    resp =  requests.get("https://cicadmin.cartodb.com/api/v2/sql?q=select%20municipio,%20distrito,%20seccion%20from%20public.nl_secciones_electorales%20where%20ST_Within%28ST_GeomFromText%28%27POINT%28" + str(long)+"%20" + str(lat)+"%29%27,%204326%29,%20the_geom%29").json()
    return resp["rows"][0]["seccion"]
```

```python
def convert_string_to_ageb(string_text):
    string_text = string_text.replace("(","")
    string_text = string_text.replace(")","")
    string_list = string_text.split(", ")
    lat = string_list[0]
    long = string_list[1]
    print(lat, long)
    try:
        ageb = get_ageb(long,lat)
    except:
        ageb = "NaN"
    print(ageb)
    return ageb
```

```python
Reportes_by_AGEB = Reportes_Seguridad_CIC_AGEB["AGEB"].value_counts()
```

```python
def stringify_ageb(ageb_as_int):
    string_ageb = str(ageb_as_int)
    string_ageb_fixed = (4-len(string_ageb))*"0"+string_ageb
    return string_ageb_fixed
```

```python
def get_indicator_by_ageb(ageb_string,indicator):
    try:
        row_with_data = Indices_AGEB_por_AGEB.loc[Indices_AGEB_por_AGEB['AGEB'] == ageb_string]
        get_indicator_value = row_with_data.iloc[0][indicator]
    except:
        get_indicator_value = "NaN"
    return get_indicator_value
```

```python
indicator_list = ["POBTOT","VPH_S_ELEC","VPH_AGUAFV","VPH_NODREN","VPH_SNBIEN"]
```

```python
def build_data_frame_with_indicators(reportes_by_ageb,indicator_list):
    base_list = []
    for ageb_values in reportes_by_ageb.iteritems():
        AGEB = stringify_ageb(ageb_values[0])
        Reportes = ageb_values[1]
        ageb_dict = { "ageb": AGEB, "reportes": Reportes}
        for indicator in indicator_list:
            ageb_dict[indicator] = get_indicator_by_ageb(AGEB,indicator)
        base_list.append(ageb_dict)
        ageb_dict = {}
    return base_list
```

```python
list_indicators_reports = build_data_frame_with_indicators(Reportes_by_AGEB,indicator_list)
```
